```python
# ---- CUSTOMER ----
import pandas as pd
import random
from datetime import datetime, timedelta

# -------------------------
# Configuration
# -------------------------

CITIES = [
    "Mumbai",
    "Delhi",
    "Bangalore",
    "Hyderabad",
    "Pune",
    "Surat",
    "Ahmedabad",
    "Chennai"
]

GENDERS = ["Male", "Female"]



# -------------------------
# Generate Customers
# -------------------------

customers = []

MALE_NAMES = [
    "Arya",
    "Krishna",
    "Vivan",
    "Harsh",
    "Rishi",
    "Arjun",
    "Vihaan",
    "Kishan",
    "Aditiya"

    
]

FEMALE_NAMES = [
    "Radha",
    "Isha",
    "Mudra",
    "Siya",
    "Pranjal",
    "Ayushi",
    "Mahek",
    "Dimple"
]
for i in range(1, 1001):

    customer_id = f"C{i:04d}"

    gender = random.choice(["Male", "Female"])

    if gender == "Male":
        first_name = random.choice(MALE_NAMES)
    else:
        first_name = random.choice(FEMALE_NAMES)
    
    city = random.choice(CITIES)

    age = random.randint(18, 60)

    gender = random.choice(GENDERS)

    signup_date = (
        datetime.today()
        - timedelta(days=random.randint(1, 730))
    ).date()

    customers.append([
        customer_id,
        first_name,
        city,
        age,
        gender,
        signup_date
    ])

# -------------------------
# DataFrame
# -------------------------

customers_df = pd.DataFrame(
    customers,
    columns=[
        "customer_id",
        "customer_name",
        "city",
        "age",
        "gender",
        "signup_date"
    ]
)

# -------------------------
# Save CSV
# -------------------------

customers_df.to_csv(
    "customers.csv",
    index=False
)

print("Customers dataset created successfully!")
print(customers_df.head())
```

```python
# ---- RESTAURANTS ----
import pandas as pd
import random
from datetime import datetime, timedelta


# -------------------------
# Generate Restaurants
# -------------------------
CITIES = [
    "Mumbai",
    "Delhi",
    "Bangalore",
    "Hyderabad",
    "Pune",
    "Surat",
    "Ahmedabad",
    "Chennai"
]


restaurants = []

RESTAURANT_NAMES = [
    "Curry Cartel",
    "The Midnight Masala",
    "Naan Stop",
    "Wok & Tikka",
    "The Green Chutney",
    "Tadka Town",
    "Dal Roti Wala",
    "The Chaat Chart"
]

CUISINE_TYPE = [
    "North Indian",
    "South Indian",
    "Chinese",
    "Pizza",
    "Biryani",
    "Nachos",
    "Sizzler"
    "Italian",
    "Mexican",
    "Gujarati"
]


rating_bucket = random.choices(
    population=[
        "low",
        "medium",
        "good",
        "excellent"
    ],
    weights=[5, 20, 50, 25],
    k=1
)[0]

for i in range(1, 101):

    restaurant_id = f"R{i:04d}"
    
    restaurant_names = random.choice(RESTAURANT_NAMES)    
    
    city = random.choice(CITIES)
    
    cuisine = random.choice(CUISINE_TYPE)
    
    if rating_bucket == "low":
        rating = round(random.uniform(3.0, 3.5), 1)

    elif rating_bucket == "medium":
        rating = round(random.uniform(3.5, 4.0), 1)

    elif rating_bucket == "good":
        rating = round(random.uniform(4.0, 4.5), 1)

    else:
        rating = round(random.uniform(4.5, 5.0), 1)
    
    partner_since = (
        datetime.today()
        - timedelta(days=random.randint(1, 1095))
    ).date()

    restaurants.append([
        restaurant_id,
        restaurant_names,
        city,
        cuisine,
        rating_bucket,
        partner_since
    ])

# -------------------------
# DataFrame
# -------------------------

restaurants_df = pd.DataFrame(
    restaurants,
    columns=[
        "restaurant_id",
        "restaurant_name",
        "city",
        "cuisine",
        "rating",
        "partner_since"
    ]
)

# -------------------------
# Save CSV
# -------------------------

restaurants_df.to_csv(
    "restaurants.csv",
    index=False
)

print("Restaurant dataset created successfully!")
print(restaurants_df.head())
```

```python
# ---- CUSTOMER ----
import pandas as pd
import random
from datetime import datetime, timedelta

# -------------------------
# Configuration
# -------------------------

CITIES = [
    "Mumbai",
    "Delhi",
    "Bangalore",
    "Hyderabad",
    "Pune",
    "Surat",
    "Ahmedabad",
    "Chennai"
]


# -------------------------
# Generate Customers
# -------------------------

partners = []

PARTNER_NAMES = [
    "Rajesh",
    "Suresh",
    "Ramesh",
    "Kushal",
    "Bajraj",
    "Shiv",
    "Vihaan",
    "Vikram",
    "sanjay"

]


for i in range(1, 51):

    partner_id = f"P{i:04d}"

    partner_name = random.choice(PARTNER_NAMES)
    
    city = random.choice(CITIES)

    vehicle_type = random.choices(
    ["Bike", "Scooter", "Cycle"],
    weights=[60, 30, 10],
    k=1)[0]

    experience_years = random.randint(0, 10)

    joining_date = (
        datetime.today()
        - timedelta(days=random.randint(1, 1095))
    ).date()

    partners.append([
        partner_id,
        partner_name,
        city,
        vehicle_type,
        experience_years,
        joining_date
    ])

# -------------------------
# DataFrame
# -------------------------

partners_df = pd.DataFrame(
    partners,
    columns=[
        "partners_id",
        "partner_name",
        "city",
        "Vehicle_Types",
        "Experience",
        "Joining_date"
    ]
)

# -------------------------
# Save CSV
# -------------------------

partners_df.to_csv(
    "partners.csv",
    index=False
)

print("Partners dataset created successfully!")
print(partners_df.head())
```

```python
# orderid , customer_id , restuarant_id , partner_id , order_date , order_value , delivery_time_minutes , order_status , customer_rating

orders = []

for i in range(1, 10001):

    order_id = f"O{i:05d}"

    customer_id = random.choice(
        customers_df["customer_id"].tolist()
    )

    restaurant_id = random.choice(
        restaurants_df["restaurant_id"].tolist()
    )

    partner_id = random.choice(
        partners_df["partners_id"].tolist()
    )

    order_date = (
        datetime.today()
        - timedelta(days=random.randint(1, 730))
    ).date()

    bucket = random.choices(
        ["low", "medium", "high", "premium"],
        weights=[50, 35, 10, 5],
        k=1
    )[0]

    if bucket == "low":
        order_value = random.randint(100, 500)

    elif bucket == "medium":
        order_value = random.randint(501, 1000)

    elif bucket == "high":
        order_value = random.randint(1001, 1500)

    else:
        order_value = random.randint(1501, 2000)


    order_status = random.choices(
    ["Delivered", "Cancelled"],
    weights=[92, 8],
        k=1)[0]

    
    if order_status == "Delivered":
        company_commission = round(order_value * 0.20,2)

        if order_value < 500:
            delivery_fee = 40

        elif order_value < 1000:
            delivery_fee = 30

        else:
            delivery_fee = 20

        delivery_time_minutes = random.randint(20,60)

        customer_rating = round(random.uniform(1, 5),1)

    else:

        company_commission = 0

        delivery_fee = 0

        delivery_time_minutes = None

        customer_rating = None


    orders.append([
        order_id,
        customer_id,
        restaurant_id,
        partner_id,
        order_date,
        order_value,
        company_commission,
        delivery_fee,
        delivery_time_minutes,
        order_status,
        customer_rating
])


orders_df = pd.DataFrame(
    orders,
    columns=[
        "order_id",
        "customer_id",
        "restaurant_id",
        "partner_id",
        "order_date",
        "order_value",
        "company_commission",
        "delivery_fee",
        "delivery_time_minutes",
        "order_status",
        "customer_rating"
    ]
)

orders_df.to_csv(
    "orders.csv",
    index=False
)

   
```

```python
# -------------------------
# PAYMENTS TABLE
# -------------------------

payments = []

for i, row in orders_df.iterrows():

    payment_id = f"PAY{i+1:05d}"

    order_id = row["order_id"]

    payment_method = random.choices(
        ["UPI", "Credit Card", "Debit Card","COD"],
        weights=[70 , 10 , 5 , 15],
        k=1
    )[0]

    if row["order_status"] == "Cancelled":

        payment_status = "Refunded"

        payment_amount = row["order_value"]

    else:

        payment_status = random.choices(
            ["Success", "Failed"],
            weights=[97, 3],
            k=1
        )[0]

        payment_amount = (
            row["order_value"]
            + row["delivery_fee"]
        )

    payment_date = row["order_date"]

    payments.append([
        payment_id,
        order_id,
        payment_method,
        payment_status,
        payment_amount,
        payment_date
    ])

# -------------------------
# CREATE DATAFRAME
# -------------------------

payments_df = pd.DataFrame(
    payments,
    columns=[
        "payment_id",
        "order_id",
        "payment_method",
        "payment_status",
        "payment_amount",
        "payment_date"
    ]
)

# -------------------------
# SAVE CSV
# -------------------------

payments_df.to_csv(
    "payments.csv",
    index=False
)

# -------------------------
# VALIDATION
# -------------------------

print(payments_df.shape)

print(
    payments_df["payment_status"]
    .value_counts()
)

print(
    payments_df.head()
)
```
